```python
import random
personas = []
def verificar_rut(rut):
    rut = rut.replace("-", "")
    rut_numero = rut[:-1]
    rut_digito_verificador = rut[-1].upper()

    if not rut_numero.isdigit() or not rut_digito_verificador.isdigit() and rut_digito_verificador != "K":
        return False


    rut_numero = int(rut_numero)
    suma = 0
    multiplicador = 2

    while rut_numero > 0:
        suma += (rut_numero % 10) * multiplicador
        rut_numero //= 10
        multiplicador += 1
        if multiplicador > 7:
            multiplicador = 2

    digito_verificador_esperado = 11 - (suma % 11)
    if digito_verificador_esperado == 10:
        digito_verificador_esperado = "K"
    elif digito_verificador_esperado == 11:
        digito_verificador_esperado = "0"


    return rut_digito_verificador == str(digito_verificador_esperado) or rut_digito_verificador == "K"


def grabar_persona():
    rut = input("Ingrese el RUT: ")
    nombre = input("Ingrese el nombre: ")
    edad = int(input("Ingrese la edad: "))
    pais_nacimiento = input("Ingrese el país de nacimiento: ")
    ciudad_nacimiento = input("Ingrese la ciudad de nacimiento: ")

    if verificar_rut(rut) and len(nombre) >= 8 and edad >= 0:
        persona = {
            "RUT": rut,
            "Nombre": nombre,
            "Edad": edad,
            "País de Nacimiento": pais_nacimiento,
            "Ciudad de Nacimiento": ciudad_nacimiento
        }
        personas.append(persona)
        print("Datos grabados correctamente.")
    else:
        print("Datos inválidos. No se pudieron grabar.")

def imprimir_certificados():
    while True:
        print("\n--- Menú de Certificados ---")
        print("1. Certificado de Nacimiento")
        print("2. Certificado de Estado Conyugal")
        print("3. Volver al menú principal")

        opcion = input("Ingrese una opción (1-3): ")

        if opcion == "1":
            imprimir_certificado_nacimiento()
        elif opcion == "2":
            imprimir_certificado_estado_conyugal()
        elif opcion == "3":
            break
        else:
            print("Opción inválida. Intente nuevamente.")
def imprimir_certificado_nacimiento():
    rut = input("Ingrese el RUT de la persona: ")
    persona = buscar_persona(rut)

    if persona:
        print("--- Certificado de Nacimiento ---")
        print("RUT: ", persona["RUT"])
        print("Nombre: ", persona["Nombre"])
        print("País de Nacimiento: ", persona["País de Nacimiento"])
        print("Ciudad de Nacimiento: ", persona["Ciudad de Nacimiento"])
    else:
        print("No se encontró ninguna persona con ese RUT.")

def imprimir_certificado_estado_conyugal():
    rut = input("Ingrese el RUT de la persona: ")
    persona = buscar_persona(rut)

    if persona:
        estados_conyugales = ["Soltero/a", "Casado/a", "Viudo/a", "Divorciado/a"]
        estado_conyugal = random.choice(estados_conyugales)

        print("--- Certificado de Estado Conyugal ---")
        print("RUT: ", persona["RUT"])
        print("Nombre: ", persona["Nombre"])
        print("Estado Conyugal: ", estado_conyugal)
    else:
        print("No se encontró ninguna persona con ese RUT.")

def buscar_persona(rut):
    for persona in personas:
        if persona["RUT"] == rut:
            return persona
    return None


def eliminar_persona():
    rut = input("Ingrese el RUT de la persona a eliminar: ")
    nombre = input("Ingrese el nombre de la persona a eliminar: ")
    edad = int(input("Ingrese la edad de la persona a eliminar: "))


    print("Persona eliminada correctamente.")


def main():
    while True:
        print("\n--- Registro Civil - TOTEM de Autoservicio ---")
        print("1. Grabar")
        print("2. Buscar")
        print("3. Imprimir certificados")
        print("4. Eliminar")
        print("5. Salir")

        opcion = input("Ingrese una opción (1-5): ")

        if opcion == "1":
            grabar_persona()
        elif opcion == "2":
            rut = input("Ingrese el RUT de la persona a buscar: ")
            persona = buscar_persona(rut)
            if persona:
                print("Persona encontrada:")
                print("RUT:", persona["RUT"])
                print("Nombre:", persona["Nombre"])
                print("Edad:", persona["Edad"])
                print("País de Nacimiento:", persona["País de Nacimiento"])
                print("Ciudad de Nacimiento:", persona["Ciudad de Nacimiento"])
            else:
                print("No se encontró ninguna persona con ese RUT.")
        elif opcion == "3":
            imprimir_certificados()
        elif opcion == "4":
            eliminar_persona()
        elif opcion == "5":
            print("¡Hasta luego!")
            break
        else:
            print("Opción inválida. Intente nuevamente.")

main()
```
